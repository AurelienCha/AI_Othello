```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import re
import os
```

```python
def create_df():
    folder = 'Cost/'
    d = pd.DataFrame()
    for file in os.listdir(folder):
        print(file)
        arg1 = file.split('_')[0][1:]
        if arg1 == '': arg1 = 'hybrid'
        arg2 = file.split('_')[1][1:]
        if arg2 == '': arg2 = 'hybrid'

        df = pd.read_csv(folder + file)
        df.columns = ['Round', *df.columns[1:]]
        zero = pd.DataFrame([[-1, 0, '', '', 2, 2]], columns=["Round", "Time", "Color", "Move", "Black", "White"])
        df = pd.concat([zero, df])
        df["Round"] = df["Round"] + 1
        df["Depth"] = int(file[0])

        df["Cost"] = df["Color"]
        df["Cost"] = df["Cost"].replace([" X ", " O "], [arg1, arg2])
        df["Score"] = df["Color"]
        df["Score"] = df["Score"].replace([" X ", " O "], [df["Black"] - df["White"], df["White"] - df["Black"]])
        df["Score"] = pd.to_numeric(df["Score"])

        d = pd.concat([d, df], axis=0, ignore_index=True)
    return d

d = create_df()
d
```

```python
d.to_csv(('Graphs/' + 'cost.csv'))
```

```python
sns.lineplot(data=d, x="Round", y="Score", hue="Cost", hue_order=['min','hybrid','max'], palette=('blue','green','red'))

plt.title("Pawns difference depending of the cost function")
plt.ylabel('Advantage')
plt.xlabel('Round')
plt.legend(title='Cost function',loc='upper left', framealpha=0.9)
plt.savefig('./Graphs/' + 'cost' + '.png')
```
